```python
import numpy as np
import matplotlib.pyplot as plt

def generate_gaus(rgi, ok):
    ck = (1/(np.sqrt(2*np.pi)*ok))**3
    fc = 0.5*(np.cos(np.pi*rgi/Rc)+1)
    unnorm = np.exp(-rgi**2/2/ok**2)
    return unnorm*fc

# Constants
Rc = 8
rgi = np.linspace(0,Rc,25)
ok_list = np.linspace(0.75, 8, 20)
for ok in ok_list:
    plt.plot(rgi, generate_gaus(rgi, ok))
plt.xlabel("Distance from ref. point")
```
