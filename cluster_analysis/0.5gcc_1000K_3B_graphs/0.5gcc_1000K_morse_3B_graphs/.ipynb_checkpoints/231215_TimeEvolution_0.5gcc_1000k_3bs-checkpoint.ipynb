```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.decomposition import PCA
from tqdm import tqdm
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
# Create an empty list to store tuples of data rows and corresponding i values
data_tuples = []

# Specify the file numbers you want to include
file_numbers = [50]

# Loop over the specified file numbers
for i in file_numbers:
    # Generate the filename
    filename = f'00{i}.3b_clu-s.txt'
    
    try:
        # Read the data from the file
        data = np.loadtxt(filename)
        data = np.sort(data, axis=1)
        
        # Loop over each line in the data and add it to the list as a tuple (data row, i value)
        for line in data:
            data_tuples.append((line, i))
    except FileNotFoundError:
        print(f"File not found: {filename}")

# Print the list of tuples
# print("List of Tuples:")
# for data_row, i_value in data_tuples:
#     print(f"{data_row}: {i_value}")

# Extract the first element of each tuple using a list comprehension
concat_data = np.array([t[0] for t in data_tuples])
# Extract the second element of each tuple using a list comprehension
statepoints = np.array([t[1] for t in data_tuples])

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the data points
ax.scatter(concat_data[:, 0], concat_data[:, 1], concat_data[:, 2], c=statepoints, marker='o', edgecolors='black')

# Set labels for each axis
ax.set_xlabel('Side A')
ax.set_ylabel('Side B')
ax.set_zlabel('Side C')

# Set axis limits to range from -1 to 1
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

# Plot red lines along the x, y, and z axes
ax.plot([-1, 1], [0, 0], [0, 0], c='red', linestyle='--', linewidth=2)  # X-axis
ax.plot([0, 0], [-1, 1], [0, 0], c='red', linestyle='--', linewidth=2)  # Y-axis
ax.plot([0, 0], [0, 0], [-1, 1], c='red', linestyle='--', linewidth=2)  # Z-axis

# Set the title of the plot
ax.set_title('Frame 50 (Start)')

# Show the plot
plt.show()
```
